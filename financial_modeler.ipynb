```python
def find_different_patterns(array):
    unique_patterns = set()

    # Convert the binary array to a string
    binary_string = ''.join(str(bit) for bit in array)

    # Find patterns of different lengths
    for pattern_length in range(1, len(array) + 1):
        for i in range(len(array) - pattern_length + 1):
            pattern = binary_string[i:i+pattern_length]
            unique_patterns.add(pattern)

    return unique_patterns


import random

# Generate an array with random binary values
array = [random.randint(0, 1) for _ in range(10)]

# Find the different types of patterns in the array
different_patterns = find_different_patterns(array)

print("Different types of patterns:", different_patterns)

# melt up. melt down. random
```

```python
import plotly.graph_objects as go
import numpy as np

# Initial investment
P = 11000

# Daily interest rate (5% per day)
r = 0.05

# Compounds per day
n = 1

# Target amounts
target = 1000000  # Stop chart at $1,000,000

# Array for plotting
days = []
amounts = []

# Days to reach each $100,000 increment
days_to_reach = []

# Calculate the amount for each day
for day in range(1, 1000):  # We're checking for up to 1000 days, adjust this as needed
    A = P * (1 + r/n)**(n*day)
    if A > target:
        break
    days.append(day)
    amounts.append(A)
    if len(days_to_reach) < A // 100000:
        days_to_reach.append(day)

# Create the plot
fig = go.Figure()

# Set color thresholds for every $100,000
colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf']
for i in range(int(target // 100000)):
    idx = np.logical_and(np.array(amounts) >= i * 100000, np.array(amounts) < (i + 1) * 100000)
    fig.add_trace(go.Scatter(x=np.array(days)[idx], y=np.array(amounts)[idx], mode='lines', line=dict(color=colors[i % len(colors)])))

# Add annotations
for i, day in enumerate(days_to_reach, start=1):
    if i == 1:
        fig.add_annotation(
            x=day, y=i*100000,
            text=f"Days to reach ${i*100000:,}: {day}",
            showarrow=False,
            yshift=10)
    else:
        fig.add_annotation(
            x=day, y=i*100000,
            text=f"Days to reach ${i*100000:,}: {day - days_to_reach[i - 2]}",
            showarrow=False,
            yshift=10)

fig.update_layout(title='Compound Interest over Time', xaxis_title='Days', yaxis_title='Amount($)',
                  autosize=False, width=800, height=500, margin=dict(l=50, r=50, b=100, t=100, pad=4))
fig.show()
```
